````python
# Find a string
def count_substring(string, sub_string):
    count = 0
    for i in range(0, len(string)):
        if string[i:i+len(sub_string)] == sub_string:
            count += 1
    return count
````

````python
# Merge the Tools! (Medium)
def merge_the_tools(string, k):
    from collections import defaultdict
    for i in range(0, len(string), k):
        target_s = string[i:i+k]
        check = defaultdict(lambda:0)
        current_s = ""
        for elem in target_s:
            if check[elem] == 0:
                current_s += elem
                check[elem] += 1
        print("".join(current_s))
````

````python
# Minion

def minion_game(string):
    stuart = 0
    kevin = 0
    for i in range(0, len(string)):
        if string[i] not in "AEIOUaeiou":
            stuart += len(string) - i
        if string[i] in "AEIOUaeiou":
            kevin += len(string) - i
    if stuart > kevin:
        result = f'Stuart {stuart}'
    elif stuart < kevin:
        result = f'Kevin {kevin}'
    else: 
        result = 'Draw' 
    print(result)
````

````python
# minion game original code - 효율 안 좋음
def minion_game(string):
    stuart = list()
    kevin = list()
    for i in range(0, len(string)):
        if string[i] not in "AEIOUaeiou":
            for j in range(i+1, len(string)+1):
                stuart.append(string[i:j])

    for i in range(0, len(string)):
        if string[i] in "AEIOUaeiou":
            for j in range(i+1, len(string)+1):
                kevin.append(string[i:j])
    if len(stuart) > len(kevin):
        result = f'Stuart {len(stuart)}'
    elif len(stuart) < len(kevin):
        result = f'Kevin {len(kevin)}'
    else: 
        result = 'Draw' 
    print(result)
````
