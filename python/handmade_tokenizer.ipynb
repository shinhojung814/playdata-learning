```python
import re
from collections import Counter

year_month_days = [
    "20191223",
    "20191224",
    "20191225",
    "20191226",
    "20191227",
    "20191228",
    "20191229",
    "20191230",
]

word_list = []

for year_month_day in year_month_days:
    f = open(f"northkoreanews_20191223_30/{year_month_day}_북한방송_주요내용.txt", 'r', encoding='UTF8')
    lines = f.readlines()
    try:
        for line in lines[2:]:
            if re.search("대\w동향",line):
                pass
            else:
                i = 0
                if line.startswith("o") or line.startswith("*"):
                    line = line[2:]
                while i < len(line):
                    current_word = ""
                    for j in range(i, len(line)):
                        if line[j] in ['인', '에', '의', '와', '"', "을", "를", "로"]:
                            pass
                        elif line[j] in [' ', '(', ')', ',', '.', '-']:
                            if current_word:
                                word_list.append(current_word)
                            break
                        else:
                            current_word += line[j]
                    i = j + 1
    except Exception as e:
            print("Exception: ", e)
    f.close()

word_counter = Counter(word_list)
word_counter.most_common
```
